```python
#IMPORTS
import urllib3
import re


#FUNCTIONS
#toString()


#DATA FIELDS
link = "https://www.metacritic.com/browse/movies/score/metascore/year/filtered?year_selected=2010&sort=desc&view=detailed"
numberOfMovies = 100

#REGULAR EXPRESSIONS
#compile regular expressions into pattern objects
movieNumberREGEX = re.compile(r'\<span class="title numbered">*\s+(\d*)')   #<-matches: 100 items
movieTitleREGEX = re.compile(r'class="title"><h3>(.*?)<\/h3>')     #r-flag transforms the following string into a string literal, removing the need for the additional backslashes. 
movieDateOfReleaseREGEX = re.compile(r'\s*<span>(\S+ \d+, \d+)')   #<-matches: 100 items
movieDescriptionREGEX = re.compile(r'<div class="summary">\n\s+(.*)')
movieMetacriticScoreREGEX = re.compile("<span class=\"title\">Metascore:<\\/span>\\s+<a class=\"metascore_anchor\" href=\"\\/movie\\/.*?\\/critic-reviews\">\s+<div class=\"metascore_w large movie positive\">(.*?)<\\/div>")
movieImageREGEX = re.compile(r'<a href=\"\/.*?<img src=\"(https:\/\/.*?)\" alt=\"')  



#
#  PULL DOWN DATA FROM METACRITIC
#
#

# Construct a python request pool
http = urllib3.PoolManager()

# Initiate a web request
r = http.request('GET', link, headers={'User-Agent': 'Mozilla/5.0'})

# Convert the result from raw bytes to text
datastring = str(r.data, "utf-8")

# Check whether it succeeded and how much data we got from the site
print(r.status)
#print(len(datastring))
```

```python
# Extract Data from HTML datastring

#Search for number of movie
list_movieNumbers = re.findall(movieNumberREGEX, datastring)
print(len(list_movieNumbers))
#print(list_movieNumbers)

#Extract list of movie titles
list_movieTitles = re.findall(movieTitleREGEX, datastring)
#print(list_movieTitles)
print(len(list_movieTitles))

#Extract list of release dates
list_releaseDates = re.findall(movieDateOfReleaseREGEX, datastring)
print(len(list_releaseDates))
#print(list_releaseDates)

#Extract list of movie summaries
list_movieDescriptions = movieDescriptionREGEX.findall(datastring)
print(len(list_movieDescriptions))
#print(list_movieDescriptions)

#Extract list of movie images
list_movieImages = movieImageREGEX.findall(datastring)
print(len(list_movieImages))
#print(list_movieImages)

#Extract list of metacritic scores
list_movieMetaCriticScores = movieMetacriticScoreREGEX.findall(datastring)
print(len(list_movieMetaCriticScores))
#print(list_movieMetaCriticScores)
```

```python
#
#
#   PUTTING IT ALL TOGETHER
#   Printed output of each movie 
#
#
#

for i in range(len(list_movieNumbers)):

    print(f"{list_movieTitles[i]}  Movie#{list_movieNumbers[i]}")
    print(f"{list_movieTitles[i]}  Released: {list_releaseDates[i]}")
    print(f"{list_movieTitles[i]}  Metacritic Score: {list_movieMetaCriticScores[i]}")
    print(f"{list_movieTitles[i]}  Image: {list_movieImages[i]}")
    print(f"{list_movieTitles[i]}  Summary: \n\t{list_movieDescriptions[i]}")

    print("---")
```
